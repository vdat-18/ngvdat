```python
print(df_combined_all['Counter_Rating'].unique())
```

```python
def convert_to_thousands(value):
    if isinstance(value, str):  # Check if the value is a string before applying lower()
        value = value.lower()
        if 'k+' in value:
            value = value.replace('k+', '')
        elif 'k' in value:
            value = value.replace('k', '')
        else:
            return int(value)  # Return the integer value if no 'k' is present

        return int(float(value) * 1000)  # Convert to thousands if 'k' was present
    else:
        return value  # Return the value as is if it's not a string
```

```python
df_combined_all['Counter_Rating'] = df_combined_all['Counter_Rating'].apply(convert_to_thousands)
```

```python
import pandas as pd

def evaluate_value(value):
    # if pd.notna(value):  # Kiểm tra giá trị không phải là NaN
        value = eval(str(value)) if isinstance(value, str) else value
        if isinstance(value, dict) and 'content' in value:  # Kiểm tra giá trị là một từ điển và có chứa trường 'content'
            return value['content']
        return value

# Áp dụng hàm cho mỗi phần tử trong DataFrame
df_copy_comments = df_copy_comments.applymap(evaluate_value)

print(df_copy_comments)
```

```python
from openpyxl.cell.cell import ILLEGAL_CHARACTERS_RE

def remove_illegal_chars(value):
    """Removes illegal characters from a string."""
    if isinstance(value, str):
        return ILLEGAL_CHARACTERS_RE.sub(r'', value)
    else:
        return value

# Apply the function to every cell in the DataFrame
df_output = df_output.applymap(remove_illegal_chars)

df_output.to_excel('/content/drive/My Drive/Colab Notebooks/NCKH (Ph. Danh Hường)/Data thô/comments data.xlsx')
```

```python
def normalize_text(sublist):
    normalize_texts = []
    for sentence in sublist:
        normalize_text = text_normalize(sentence)
        normalize_texts.append(normalize_text)
    return normalize_texts

df_copy = df_copy.apply(normalize_text)
```

```python
def tokenize_text(sublist):
    tokenize_texts = []
    for sentence in sublist:
          tokenize_text = sent_tokenize(sentence)
          tokenize_texts.append(tokenize_text)
    return  tokenize_texts

df_copy = df_copy.apply(tokenize_text)
```

```python
df_copy = df_copy.apply(lambda x: None if x == ['None'] else x).dropna()
```

Cần xử lý đoạn dưới
Xử lý 'None'
- Cần đưa về None nếu là None
- Hiện tại None đang đưa về positive

```python
def sentiment_text(sublist):
    sentiment_texts = []

    for sentence in sublist:
        sen_texts = []
        for text in sentence:
            if text and text != ['nan']:
                positive_count = negative_count = none_count = 0

                for index in text:
                    if index != '.':
                        senti_index = sentiment(index)

                        if senti_index == 'positive':
                            positive_count += 1
                        elif senti_index == 'negative':
                            negative_count += 1
                        # Check if senti_index is not None and not empty before calling strip()
                        elif senti_index is not None and not senti_index.strip():
                            none_count += 0.1

                if positive_count > negative_count:
                    sen_texts.append('positive')
                elif positive_count < negative_count:
                    sen_texts.append('negative')
                elif none_count > max(positive_count, negative_count):
                    sen_texts.append('None')
                else:
                    sen_texts.append('positive')

        sentiment_texts.append(sen_texts)

    return sentiment_texts

# Áp dụng hàm tính toán cảm xúc vào từng dòng của cột text_column
sentiment_text(df_copy)
```

```python
results = sentiment_text(df_copy)
data = []
for result in results:
  data.append(result)
data = pd.DataFrame(data)
data = data.apply(lambda x: ', '.join(x.astype(str)), axis=1)
```
